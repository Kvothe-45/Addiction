```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_teen_phone = pd.read_csv("BDD_initial/teen_phone_addiction_dataset.csv", sep=",", index_col=0)
data_teen_phone.head()
```

```python
sns.scatterplot(data=data_teen_phone, x="Daily_Usage_Hours", y="Sleep_Hours",hue="Phone_Usage_Purpose")
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(
    data=data_teen_phone,
    x='Daily_Usage_Hours',
    y='Sleep_Hours',
    hue='Phone_Usage_Purpose',
    dodge=True,
    alpha=0.9,
    palette='Set1'
)

plt.title('Age vs Cigarettes par jours par sexe')
plt.xlabel('Smokes per Day')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = sns.regplot(data=data_teen_phone, x="Time_on_Social_Media", y="Sleep_Hours", ci=95, scatter_kws={"alpha":0.6})
ax.set_title("RÃ©gression linÃ©aire (IC 95%) â€” iris")
plt.show()
```

```python
ax = sns.barplot(data=data_teen_phone, x="Phone_Usage_Purpose", y="Daily_Usage_Hours", ci=95, estimator=np.mean, hue="Anxiety_Level")
ax.set_title("temps d'usage du tÃ©lÃ©phone par application, \nniveaux d'anxiÃ©tÃ© et nombre d'heures")
```
