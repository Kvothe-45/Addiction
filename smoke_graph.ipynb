```python
%pip install seaborn
%pip install pandas
%pip install numpy
%pip install matplotlib

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_smoke_drink = pd.read_csv("BDD_initial/addiction_population_data.csv", sep=",", index_col=0)
data_smoke_drink.head()
```

```python
ordre = ['Secondary', 'Primary', 'Postgraduate', 'University', 'College',
       'High School']
fig, ax = plt.subplots()
sns.violinplot(
    data=data_smoke_drink,
    x="education_level",
    y="sleep_hours",
    order=ordre,
    inner="quartile",
    ax=ax,
)
ax.set_title("Heures de sommeil par niveau acadmique")
ax.set_xlabel("Niveau acadmique")
ax.set_ylabel("Heures sommeils par nuit")
plt.show()

"""
Ce graphique en violon montre la distribution des heures de sommeil par nuit en fonction du niveau acadmique. 
Chaque "violon" reprsente la distribution des heures de sommeil pour un niveau acadmique spcifique, avec les quartiles indiqus  l'intrieur. 
On peut constater que le niveau acadmique ne semble pas avoir d'impact significatif sur le nombre d'heures de sommeil, car tous les quartiles des violons  plots se situent dans une plage similaire, avec une mdiane aux alentours de 6,2 heures de sommeil.
"""
```

```python
fig = plt.figure()
ax = sns.barplot(data=data_smoke_drink, x="education_level", y="annual_income_usd", hue="addict_smoke", estimator=np.mean, ci=95)
ax.set_title("Revenu annuel moyen par niveau acadmique et addiction au tabac")
plt.show()

"""
Ce graphique  barres montre le revenu annuel moyen en fonction du niveau acadmique et de leurs addictions au tabac. 
On ne peut cependant pas visualiser une quelconque relation entre ces variables, le revenu annuel semble rester stable entre le niveau acadmique et l'addiction des personnes. 
"""
```

```python
fig = plt.figure()
ax = sns.regplot(data=data_smoke_drink, x="bmi", y="sleep_hours", ci=95, scatter_kws={"alpha":0.6})
ax.set_title("Rgression linaire (IC 95%)  iris")
plt.show()
"""
Ce graphique de rgression linaire montre l'absence de relation linaire (ou d'autre relation) entre les heures de sommeil et le BMI.
"""
```

```python
fig = plt.figure()
ax = sns.scatterplot(data=data_smoke_drink, x="annual_income_usd", y="sleep_hours", hue="addict_smoke")
plt.show()

"""
Ce nuage de points montre l'absence de relation entre les heures de sommeil, le revenu annuel et l'addiction. 
Les points sont disperss de manire alatoire, sans tendance claire, ce qui suggre qu'il n'y a pas de corrlation significative entre ces variables.

"""
```

```python


plt.figure(figsize=(10, 6))
sns.stripplot(
    data=data_smoke_drink,
    x='smokes_per_day',
    y='age',
    hue='gender',
    dodge=True,
    alpha=0.9,
    palette='Set1'
    )

plt.title('Age vs Cigarettes par jours par sexe')
plt.xlabel('Smokes per Day')
plt.ylabel('Age')
plt.grid(True)
plt.show()

"""
Ce stripplot nous permet de constater encore une fois une absence apparente de relation entre le nombre de cigarettes fumes par jour, l'ge et le genre. 
Les variables semblent disperses de manire alatoire (les classes tant spares  l'aide de l'argument dodge) 
"""
```
