```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_smoke_drink = pd.read_csv("BDD_initial/addiction_population_data.csv", sep=",", index_col=0)
data_smoke_drink.head()
```

```python
ordre = ['Secondary', 'Primary', 'Postgraduate', 'University', 'College',
       'High School']
fig, ax = plt.subplots()
sns.violinplot(
    data=data_smoke_drink,
    x="education_level",
    y="sleep_hours",
    order=ordre,
    palette="Set2",
    inner="quartile",
    ax=ax,
)
ax.set_title("Heures de sommeil par niveau acadÃ©mique")
ax.set_xlabel("Niveau acadÃ©mique")
ax.set_ylabel("Heures sommeils par nuit")
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = sns.barplot(data=data_smoke_drink, x="education_level", y="annual_income_usd", hue="addict_smoke", estimator=np.mean, ci=95)
ax.set_title("Revenu annuel moyen par niveau acadÃ©mique et addiction au tabac")
plt.show()
```

```python
fig = plt.figure()
ax = sns.regplot(data=data_smoke_drink, x="bmi", y="sleep_hours", ci=95, scatter_kws={"alpha":0.6})
ax.set_title("RÃ©gression linÃ©aire (IC 95%) â€” iris")
plt.show()
```
